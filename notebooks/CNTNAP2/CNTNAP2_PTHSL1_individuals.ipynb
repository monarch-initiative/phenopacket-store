# CNTNAP2
[Pitt-Hopkins-like syndrome-1 (PTHSL1)](https://omim.org/entry/610042) is caused by homozygous or compound heterozygous mutation in the CNTNAP2 gene.

```python
from pyphetools.creation import TemplateImporter, Moi
from pyphetools.visualization import IndividualTable, QcVisualizer
from IPython.display import display, HTML
import pyphetools
print(f"Using pyphetools version {pyphetools.__version__}")
```

```python
template = "input/CNTNAP2_PTHSL1_individuals.xlsx"
created_by = "0000-0002-0736-9199"
```

```python
del_set = {"deletion c.98-?_402+? Encompassing exon 2-3", 
           "7q35(147,651,818-147,683,766)x1 encompassing exon 15",
           "7q36.1(148,071,316-148,080,632)x1 encompassing exon 22",
           "deletion c.98-?_402+? Encompassing exon 2-3",
           "deletion c.98-?_1348+? Encompassing exons 2-8",
           "del exons 2,3,4",
           "chr7:(146.079.333-146.194.785)x1",
           "chr7:(146.144.267-146.374.539)x1",
           "chr7(147.702.165-148.378.711)x1",
           "del exon 2",
           "95.4 kb deletion intron 1",
           "220 kb deletion",
           "246 kb deletion in 7q35",
           "deletion of exon 4",
           "270 kb deletion across exons 2-3",
           "del hg38; Chr7:146,711,006-146,914,175",
           "del exon 2-3",
           "del exon 1 (56kb)",
           "del exon 9-10", 
           "del exon 4-20 (1.23 Mb)",
           "del exon 1",
           "DEL exon 3 (76.8 kb)",
           "deletion of exons 5-8",
           "deletion of exons 2-9"
           }
dup_set = {"duplication in intron 1 and exon 2", "dup exons 2–11"}
timporter = TemplateImporter(template=template,  created_by=created_by)
individual_list, cvalidator = timporter.import_phenopackets_from_template(deletions=del_set, duplications=dup_set, leniant_MOI=True)
```

```python
qc = QcVisualizer(cohort_validator=cvalidator)
display(HTML(qc.to_summary_html()))
```

```python
table = IndividualTable(cvalidator.get_error_free_individual_list())
display(HTML(table.to_html()))
```
