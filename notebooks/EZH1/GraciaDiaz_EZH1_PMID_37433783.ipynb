```python
# Finally, the only non-consanguineous case is a child that carries two variants in compound heterozygosity: 
#a large deletion covering exon 8â€“12 of EZH1 inherited from the mother 
# (c.[664â€‰+â€‰1_665-1]_[1401â€‰+â€‰1_1402-1]del; p.?) 
# and a de novo splice variant affecting the exon 10 splice acceptor site (c.932-1G>A; p.?)
mutation_d = {
    "p.R406H (heterozygous)": "c.1217G>A", # VCV000828189.1
    "p.E438D (heterozygous)":"c.1314A>T",  # NP_001982.2:p.(Glu438Asp)
    "p.K612M (heterozygous)":"c.1835A>T",  # NP_001982.2:p.(Lys612Met)
    "p.A678G (heterozygous)": "c.2033C>G", # VCV000977755.3
    "p.R728G (heterozygous)":"c.2182A>G",  # NP_001982.2:p.(Arg728Gly) - inferred"
    "p.Q731E (heterozygous)": "c.2191C>G", # VCV002570596.1
    "p.L735F (heterozygous)": "c.2203C>T", # VCV002570597.1
    "Stop gain (p.R258X) homozyogus": "c.772C>T", # "c.772C>T; p.R258X
    "c.932-1G>A": "c.932-1G>A",
    "a large deletion covering exon 8â€“12 of EZH1-c.[664+1_665-1]_[1401+1_1402-1]del":"a large deletion covering exon 8â€“12 of EZH1-c.[664+1_665-1]_[1401+1_1402-1]del",
    "splice":  "c.1401+3_1403del",                 # Variant validator
    "deletion":  "c.664+2_665-1del",               # Variant validator
    "Stop gain (p.E485X) homozyogus": "c.1453G>T",  # VCV002570598.1"
    "Stop gain (p.Q413X) homozygous": "c.1237C>T",  # "c.1137C>T; p.Q413X - in original pub. Author corrected via email 17/7/23"]  
    "na":"na"
}
# Variant validator
# NM_001991.5:c.1401+3_1403del
# NM_001991.5:c.664+2_665-1del
# Create a column with the above variants
def get_var(description):
    if description in mutation_d:
        return mutation_d.get(description)
    else:
        raise ValueError(f"Could not map {description}")

df['allele_1'] = df['mutation_1'].apply(lambda x: get_var(x))
df['allele_2'] = df['mutation_2'].apply(lambda x: get_var(x))
```

```python
EZH1_transcript = "NM_001991.5"
EZH1_id = "HGNC:3526"
vman = VariantManager(df=df, individual_column_name="individual_id", allele_1_column_name="allele_1",
                      allele_2_column_name="allele_2", gene_symbol="EZH1", gene_id=EZH1_id,
                      transcript=EZH1_transcript)
```

```python
sexMapper = SexColumnMapper(male_symbol='Male', female_symbol='Female', column_name='Gender')
#sexMapper.preview_column(df)
encoder = CohortEncoder(df=df, 
                        hpo_cr=hpo_cr, 
                        column_mapper_list=column_mapper_list, 
                        individual_column_name="individual_id", 
                        age_of_onset_mapper=AgeColumnMapper.not_provided(), 
                        age_at_last_encounter_mapper=AgeColumnMapper.not_provided(),
                        sexmapper=sexMapper,
                        metadata=metadata)
omim_label = "EZH1-related neurodevelopmental disorder"
omim_id = "OMIM:601674"
disease = Disease(disease_id=omim_id, disease_label=omim_label) # Note, no OMIM entryyet
encoder.set_disease(disease)
```

```python
individual_list = encoder.get_individuals()
```

```python
individual_to_alleles_d =  vman._individual_to_alleles_d
variant_d = vman.get_variant_d()
for i in individual_list:
    allele_list = individual_to_alleles_d.get(i.id)
    if len(allele_list)==1:
        var = variant_d.get(allele_list[0])
        var.set_heterozygous()
        i.add_variant(var)
    elif len(allele_list) == 2:
        allele_1 = allele_list[0]
        allele_2 = allele_list[1]
        var1 = variant_d.get(allele_1)
        if allele_1 == allele_2:
            var1.set_homozygous()
            i.add_variant(var1)
        else:
            var2 = variant_d.get(allele_2)
            var1.set_heterozygous()
            i.add_variant(var1)
            var2.set_heterozygous()
            i.add_variant(var2)
    else:
        raise ValueError("Invalid allele list length - should never happen")
```
