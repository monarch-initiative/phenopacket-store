<h1>GLI3: Polydactyly, postaxial, types A1 and B</h1>
<p>[Polydactyly, postaxial, types A1 and B 	OMIM:174200](https://omim.org/entry/174200) is caused by monoallelic variants in the GLI3 gene.</p>
<p>The clinical data was derived from <a href="https://pubmed.ncbi.nlm.nih.gov/22428873/"target="__blank">Al-Qattan MM. A novel frameshift mutation of the GLI3 gene in a family with broad thumbs with/without big toes, postaxial polydactyly and variable syndactyly of the hands/feet. Clin Genet. 2012 Nov;82(5):502-4 PMID:22428873</a>.<p>

```python
from pyphetools.creation import TemplateImporter
from pyphetools.visualization import QcVisualizer, IndividualTable
from IPython.display import HTML, display
import pyphetools
print(f"pyphetools version {pyphetools.__version__}")
```

```python
hp_json = "../hp.json"
created_by="ORCID:0000-0002-0736-9199"
template = "input/GLI3_PAPA1_individuals.xlsx"
timporter = TemplateImporter(hp_json=hp_json, template=template, created_by=created_by)
```

```python
individuals, cvalidator = timporter.import_phenopackets_from_template()
```

```python
qc = QcVisualizer(cohort_validator=cvalidator)
display(HTML(qc.to_summary_html()))
```

```python
table = IndividualTable(individuals)
display(HTML(table.to_html()))
```

```python
PMID = "PMID:22428873"
timporter.create_hpoa_from_phenopackets(pmid=PMID, moi="Autosomal dominant", target="OMIM:174200")
```
