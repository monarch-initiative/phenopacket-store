# IKZF1
[common variable immunodeficiency-13 (CVID13)](https://omim.org/entry/616873) is caused by heterozygous mutation in the IKZF1 gene.

CVID13 is an autosomal dominant primary immunodeficiency disorder characterized by recurrent bacterial infections, mainly affecting the respiratory tract, and associated with hypogammaglobulinemia and decreased numbers of B cells.
The IKAROS gene encodes for a C2H2 zinc finger protein whose expression is restricted to hematopoietic cells and the pituitary gland [1,2]. Proteins generated from the IKAROS gene contain two separate zinc finger domains. Four zinc fingers in the amino half of the protein take part in sequence-specific DNA binding [3]. At the C-terminus of the protein there are two additional zinc fingers that do not bind DNA but which are responsible for protein-protein interaction ([Goldman et al. (2011)](https://pubmed.ncbi.nlm.nih.gov/21548011/)).

Fifteen unique HI variants have been described. Only two missense variants affecting the same amino acid (N159) have been described to act in a DN manner in humans. 
HI and DN variants result in distinct clinical and immunological phenotypes ([Santos 2020](https://pubmed.ncbi.nlm.nih.gov/32654692/)). 
Patients with de novo heterozygous dominant-negative mutations (N159S or N159T) have severe clinical and immunological phenotype compared to the patients with IKAROS haploinsufficiency.

- in-vitro analysis of the R143W mutation confirmed that it behaves in a partial dominant-negative manner on DNA binding and PC-HC localization, and opposite to the haploinsufficiency effect originally described ([PMID:33122583](https://pubmed.ncbi.nlm.nih.gov/33122583/))

```python
from pyphetools.creation import TemplateImporter, Moi
from pyphetools.visualization import IndividualTable, QcVisualizer
from IPython.display import display, HTML
import pyphetools
print(f"Using pyphetools version {pyphetools.__version__}")
```

```python
template = "input/IKZF1_CVID13_individuals.ipynb.xlsx"
created_by = "0000-0002-0736-9199"
```

```python
deletion_set = {"4.7 deletion", "16.8kb deletion"}
timporter = TemplateImporter(template=template,  created_by=created_by)
individual_list, cvalidator = timporter.import_phenopackets_from_template(deletions=deletion_set)
```

```python
qc = QcVisualizer(cohort_validator=cvalidator)
display(HTML(qc.to_summary_html()))
```

```python
table = IndividualTable(cvalidator.get_error_free_individual_list())
display(HTML(table.to_html()))
```
