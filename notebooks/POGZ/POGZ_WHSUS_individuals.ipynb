# POGZ
[White-Sutton syndrome (WHSUS)](https://omim.org/entry/616364) is caused by heterozygous mutation in the POGZ gene.

```python
from pyphetools.creation import TemplateImporter, Moi
from pyphetools.visualization import IndividualTable, QcVisualizer
from IPython.display import display, HTML
import pyphetools
print(f"Using pyphetools version {pyphetools.__version__}")
```

```python
template = "input/POGZ_WHSUS_individuals.xlsx"
created_by = "0000-0002-0736-9199"
```

```python
del_set = {"Deletion 1q21.3 encompassing whole POGZ-gene", "32 kb del in 1q21.3 (Exons 4-19)"}
timporter = TemplateImporter(template=template,  created_by=created_by)
individual_list, cvalidator = timporter.import_phenopackets_from_template(deletions=del_set)
```

```python
qc = QcVisualizer(cohort_validator=cvalidator)
display(HTML(qc.to_summary_html()))
```

```python
table = IndividualTable(cvalidator.get_error_free_individual_list())
display(HTML(table.to_html()))
```

```python
pmid = "PMID:35052493"
df = timporter.create_hpoa_from_phenopackets(pmid=pmid, mode_of_inheritance=Moi.AD, target="OMIM:616364")
df.head()
```
