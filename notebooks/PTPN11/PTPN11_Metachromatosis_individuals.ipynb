# Metachromatosis

In metachondromatosis (MC), patients develop tumors both on and within bones. These tumors are called exostoses (on the surface of bone) or endosteal tumors (within bone).

PTP1N11-relevant clinical details are not provided in relevant publicationa, but from the description we infer that each affected individual had the following:
- [Multiple enchondromatosis - HP:0005701](https://hpo.jax.org/app/browse/term/HP:0005701)
- [Multiple exostoses - HP:0002762](https://hpo.jax.org/app/browse/term/HP:0002762)
- From literature on Metachondromatosis we infer that the following features are absent in affected individuals, because they would be very likely to be observed if present and would be considered relevant given the features
of PTPN11-related Noonan syndrome: Downslanted palpebral fissures, Webbed neck, Posteriorly rotated ears, Pulmonic stenosis, Hypertelorism, Intellectual disability

```python
from pyphetools.creation import TemplateImporter
from pyphetools.visualization import IndividualTable, QcVisualizer
from IPython.display import display, HTML
import pyphetools
print(f"Using pyphetools version {pyphetools.__version__}")
```

```python
template = "input/PTPN11_Metachondromatosis_individuals.xlsx"
hp_json = "../hp.json"
created_by = "0000-0002-0736-9199"
```

```python
timporter = TemplateImporter(template=template, hp_json=hp_json, created_by=created_by)
individual_list, cvalidator = timporter.import_phenopackets_from_template()
```

```python
qc = QcVisualizer(cohort_validator=cvalidator)
display(HTML(qc.to_summary_html()))
```

```python
table = IndividualTable(cvalidator.get_error_free_individual_list())
display(HTML(table.to_html()))
```
