# LMNA FPLD2
[Familial partial lipodystrophy type 2 (FPLD2)](https://omim.org/entry/151660) is caused by heterozygous mutation in the gene encoding lamin A/C (LMNA).

- Data taken from  [Lanktree M, et al. (2007) Novel LMNA mutations seen in patients with familial partial lipodystrophy subtype 2 (FPLD2; MIM 151660). Clin Genet ;71(2):183-6. PMID:17250669](https://pubmed.ncbi.nlm.nih.gov/17250669/).

The authors write:
The clinical and biochemical attributes of the three patients are shown in Table 1. Each had typical loss of subcutaneous adipose in the limbs and gluteal region, although these findings were less pronounced for patient C, as often is the case with men. Each had insulin resistance and type 2 diabetes mellitus although none had acanthosis nigricans. Each had high triglycerides and depressed high-density lipoprotein cholesterol. No findings of other laminopathies were present in any patient.

The authors write that No findings of other laminopathies were present in any patient.

- Muscle weakness (HP:0001324)



- Data extracted from [Schmidt HH, et al. (2001) Dyslipemia in familial partial lipodystrophy caused by an R482W mutation in the LMNA gene. J Clin Endocrinol Metab. 86(5):2289-95.  PMID:11344241](https://pubmed.ncbi.nlm.nih.gov/11344241/)

The authors write:
The affected adult patients of this family characteristically present in the sequence of appearance with loss of sc fat of the extremities, muscular hypertrophy especially of the lower legs, acanthosis nigricans, severe hypertriglyceridemia, and insulin-resistant diabetes mellitus. Remarkably, affected members of this family present with severe myalgia, especially after exercise.
See also the clinical vignette.

Here, we created phenopackets for the three adults members of the family for whome detailed clinical information is available.

Additional data was curated from [Besci O, et al. Deciphering the Clinical Presentations in LMNA-related Lipodystrophy: Report of 115 Cases and a Systematic Review. J Clin Endocrinol Metab. 2024 Feb 20;109(3):e1204-e1224](https://pubmed.ncbi.nlm.nih.gov/37843397/). For curation, the following assumptions were made

- LDL cholesterol: curated as elevated for 130 mg/dL and above
- HBA1c: curated as elevated for  5.7% and above
- HDL cholesteral: reduced if 40 mg/dl or below
- trigylcerides: elevated if High: 200 mg/dL or above
- total cholesteral: elevated if 200 mg/dl or above

The patients carrying the R644C variant demonstrated an atypical FPLD2 phenotype (patients Patient 1.1-1.7), while the rest of the patients demonstrated the classic FPLD2 phenotype.
Ninety-eight percent (n = 89) of patients presented with FPLD2 phenotypes, while one had generalized lipodystrophy without progeroid features (n = 1, T10I). 
Typical FPLD2 was defined by loss of adipose tissue from the upper and lower extremities and trunk, with an excess of adipose tissue in the face, neck, and supraclavicular region. 

Atypical FPLD2 was defined when adipose tissue loss was less severe, with lipoatrophy being more evident in the legs. 

Coded as

- Loss of subcutaneous adipose tissue in limbs HP:0003635
- Loss of truncal subcutaneous adipose tissue HP:0009002 (not atypical)
- Generalized lipodystrophy HP:0009064 (Patient with T110I)



```python
from pyphetools.creation import TemplateImporter, Moi
from pyphetools.visualization import IndividualTable, QcVisualizer
from IPython.display import display, HTML
import pyphetools
print(f"Using pyphetools version {pyphetools.__version__}")
```

```python
template = "input/LMNA_FPLD2_individuals.xlsx"
created_by = "0000-0002-0736-9199"
timporter = TemplateImporter(template=template, created_by=created_by)
individual_list, cvalidator = timporter.import_phenopackets_from_template()
```

```python
qc = QcVisualizer(cohort_validator=cvalidator)
display(HTML(qc.to_summary_html()))
```

```python
table = IndividualTable(cvalidator.get_error_free_individual_list())
display(HTML(table.to_html()))
```
