# LMNA: CMD1A
[Cardiomyopathy, dilated, 1A](https://omim.org/entry/115200) is caused by heterozygous variants in LMNA.

Part of data in this template were derived from Table 1 of [Fatkin D, et al. (1999) Missense mutations in the rod domain of the lamin A/C gene as causes of dilated cardiomyopathy and conduction-system disease. N Engl J Med 341(23):1715-24. PMID:10580070](https://pubmed.ncbi.nlm.nih.gov/10580070/).


Variants were curated as follows
- Family A: Arg571Ser; C→A (1711): NM_005572.4(LMNA):c.1711C>A (p.Arg571Ser)
- Family B: Asn195Lys; C→G (585): NM_170707.4(LMNA):c.585C>A (p.Asn195Lys) 
- Family C: Leu85Arg; T→G (254): NM_170707.4(LMNA):c.254T>G (p.Leu85Arg) 
- Family D: Arg60Gly; C→G (178): NM_170707.4(LMNA):c.178C>G (p.Arg60Gly)
- Family E: Glu203Gly; A→G (608):  NM_170707.4(LMNA):c.608A>G (p.Glu203Gly)

Note that in the MANE SELECT trasncript, NM_005572.4(LMNA):c.1711C>A (p.Arg571Ser) is equivalent to NM_170707.4:c.1698+13C>A. Therefore, in the Excel file, we use the transcript NM_005572.4. This will retrieve the correct genomic coordinates.

The authors state that No family members with mutations had either joint contractures or skeletal myopathy. We therefore recorded the following terms  as excluded
- Joint contracture (HP:0034392)
- Muscle weakness (HP:0001324)
- Skeletal muscle atrophy (HP:0003202)

Additionally,  we will code the following two features as excluded (which are associated with other laminopathies but would certainly be ascertained by even a basic clinical examination)
- Progeroid facial appearance HP:0005328 (see OMIM)
- Lipodystrophy HP:0009125 (see OMIM)

In [Sébillon et al (2003)](https://pubmed.ncbi.nlm.nih.gov/12920062/), the clinical data table indicates that some individuals had skeletal muscular dystrophy with mild weakness and muscle wasting. We exclude the skeletal muscle terms in the others and report weakness and atrophy in the individuals with dystrophy (including the indivudal reported to be borderline).

```python
from pyphetools.creation import TemplateImporter
from pyphetools.visualization import IndividualTable, QcVisualizer
from IPython.display import display, HTML
import pyphetools
print(f"Using pyphetools version {pyphetools.__version__}")
```

```python
template = "input/LMNA_CMD1A_individuals.xlsx"
created_by = "0000-0002-0736-9199"
timporter = TemplateImporter(template=template,  created_by=created_by)
individual_list, cvalidator = timporter.import_phenopackets_from_template()
```

```python
qc = QcVisualizer(cohort_validator=cvalidator)
display(HTML(qc.to_summary_html()))
```

```python
table = IndividualTable(cvalidator.get_error_free_individual_list())
display(HTML(table.to_html()))
```
