# POLR1A
[hypomyelinating leukodystrophy-27 (HLD27)](https://omim.org/entry/620675) is caused by homozygous mutation in the POLR1A gene.

```python
from pyphetools.creation import TemplateImporter
from pyphetools.visualization import IndividualTable, QcVisualizer
from IPython.display import display, HTML
import pyphetools
print(f"Using pyphetools version {pyphetools.__version__}")
```

```python
template = "input/POLR1A_HLD27_individuals.xlsx"
created_by = "0000-0002-0736-9199"
```

```python
timporter = TemplateImporter(template=template,  created_by=created_by)
individual_list, cvalidator = timporter.import_phenopackets_from_template()
```

```python
qc = QcVisualizer(cohort_validator=cvalidator)
display(HTML(qc.to_summary_html()))
```

```python
table = IndividualTable(cvalidator.get_error_free_individual_list())
display(HTML(table.to_html()))
```

```python
pmid = "PMID:28051070"
timporter.create_hpoa_from_phenopackets(pmid=pmid, moi="Autosomal recessive", target="OMIM:620675")
```
